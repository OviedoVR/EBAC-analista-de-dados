1.1. Crie uma lista chamada `paises` com o pais dos usuários com mais de 30 anos. 

```
#Criando uma lista vazia de países:
paises = []

#Inicando o loop de repetições:
for dado_de_usuario in propaganda_online:
  
  #Aplicando a condição de idade:
  if dado_de_usuario['idade'] > 30:
    #Se satisfeita, adicione os dados de cada usário na lista paises:
    paises.append(dado_de_usuario['pais'])

#Imprime o resultado:
print(paises)
```

1.2. Crie uma lista chamada `leads` com a renda dos usuários que clicaram na propaganda. 

```
#Criando uma lista vazia de leads:
leads = []

#Loop de repetição para obter os dados de cada usuário:
for dado_de_usuario in propaganda_online:

  #Tentando rodar o código
  try:
    if dado_de_usuario['clicou_no_ad'] == 1:
      leads.append(dado_de_usuario['renda_area'])  
  #Lidando com o erro/exceção:  
  except KeyError:
    if dado_de_usuario.keys() == None:
      print('Chave não existente, ou seja, um usuário sem a informação "Clicou na propaganda".') 

#Imprimindo a lista de leads:
print(leads)
```

1.3. Crie uma lista chamada `cidades` com a cidade dos usuários que passaram mais de 70 segundos no site. 

```
#Criando uma lista vazia de cidades :
cidades = []

#Loop de repetição:
for dado_de_usuario in propaganda_online:
  
  #Tentando rodar o código
  try:
    if dado_de_usuario['tempo_gasto_site'] > 70:
      cidades.append(dado_de_usuario['cidade'])
  
  #Lidando com o erro/exceção:
  except TypeError:
      print('É impossível comparar uma variável do tipo "vazio" com um inteiro. Isso gera o erro "Type Error" , já que um dos dados não apresenta a informação do tempo gasto no site.')

  #Tentando solucionar:
  finally:
    if dado_de_usuario.values() == None:
      #Trocando o valor vazio para para zero:
      dado_de_usuario['tempo_gasto_site'] = 0

#Imprime a lista com as cidades:
print(cidades)
```
