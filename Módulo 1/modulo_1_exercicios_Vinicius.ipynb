## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A)
A = 3 * 320.52

#Imprindo na tela o valor de A:
print('(A):', A)

#Checando pelo Ticket Médio:
print('Ticket Médio:', A / 3)
```

```
# (B) 
B = 834.47 / 119.21

#Imprimindo na tela o valor de B:
print('(B):', int(B))

#Checando pelo Ticket Médio:
print('Ticket Médio:', 834.47 / B)
```

```
# (C)
C = 15378.12 / 5

#Imprimindo na tela o valor de C:
print('(C):', round(C,2))

#Checando pelo Valor Total de Vendas:
print('Valor Total de Vendas:', C*5)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
```

```
#Split: utiliza um delimitador para separar a string em um lista de strings:
print(cancao.split(','))

#Count: serve para contagem de determinado string:
print(cancao.count('m'))

#Replace: troca uma string por outra:
print(cancao.replace('.','!'))
```
